```python
import folium
import numpy as np
import pandas as pd
import geopandas as gpd

#import census tract shapefiles
select_tracts = ["380001","380002","378000","365002","365003","379000","377000","376000","374000","371000","367200"]
tracts = gpd.read_file("base-data/cb_2023_06_tract_500k.zip", dtype={'FIPS_11_digit':str})
tracts = tracts[tracts["TRACTCE"].isin(select_tracts)]

m = folium.Map(location=(37.94, -122.35), zoom_start=14, tiles="cartodb positron")
folium.GeoJson(tracts).add_to(m)

m
```
